# **Rain Data**

```python
df_rain
```

count nan in rain data and make it to csv file

```python
df_rain_nan=pd.DataFrame(df_rain.isna().sum())
df_rain_nan.rename(columns ={0:'rain_nan_counts'},inplace = True)
df_rain_nan=df_rain_nan.sort_values(by='rain_nan_counts')
df_rain_nan.to_csv('rain_nan_counts.csv')
```

```python
df_rain_nan.reset_index(inplace=True)
df_rain_nan.rename(columns={'index':'code'},inplace=True)
```

```python
df_rain_nan=df_rain_nan.drop(0)
df_rain_nan.to_csv('rain_nan_counts.csv')
```

# **Station Data**

```python
df_station=pd.read_excel('rain/station.xlsx')
```

```python
df_station
```

combine station data and rain nan counts data

```python
df_station_rain=df_station.merge(df_rain_nan, on='code', how='left')
df_station_rain.sort_values(by='rain_nan_counts',ignore_index=True,inplace=True)
```

```python
df_station_rain
```

```python
df_station_rain.to_csv('station_rain_nan_counts')
```

#**Temperature all years data nan counts**



```python
import glob

files = glob.glob('temp/temp/*.csv')
df_temp = pd.concat([pd.read_csv(fp) for fp in files], ignore_index=True)
```

```python
df_temp.replace(-999,np.nan,inplace=True)
```

```python
df_temp
```

```python
df_temp_nan=pd.DataFrame(df_temp.isna().sum())
df_temp_nan.rename(columns ={0:'temp_nan_counts'},inplace = True)
df_temp_nan=df_temp_nan.sort_values(by='temp_nan_counts')
df_temp_nan.reset_index(inplace=True)
df_temp_nan.rename(columns={'index':'code'},inplace=True)
df_temp_nan=df_temp_nan.drop([0,1])
df_temp_nan.to_csv('temp_nan_counts.csv')
df_temp_nan
```

```python
df_station_temp=df_station.merge(df_temp_nan, on='code', how='left')
df_station_temp.sort_values(by='temp_nan_counts',ignore_index=True,inplace=True)
```

```python
df_station_temp
```

```python
df_station_temp.to_csv('station_temp_nan_counts.csv')
```

#**combine nan**

```python
df_station_temp_rain=df_station_temp.merge(df_station_rain.loc[:,['code','rain_nan_counts']], on='code', how='left')
df_station_temp_rain['temp_rain_nan']=df_station_temp_rain['temp_nan_counts']+df_station_temp_rain['rain_nan_counts']
df_station_temp_rain.sort_values(by='temp_rain_nan')
```

```python
df_station_temp_rain.to_csv('station_temp_rain_combine.csv')
```
